```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()

items = ingredients[504:]

for item in items:

    driver.get('https://www.instacart.com/store/s?k='+item)

    time.sleep(15)
    try:
        element = driver.find_element(By.CSS_SELECTOR, 'li > span > button > div > img')
        element.click()

        time.sleep(5)

        weight = "None"
        price = "None"

        driver.execute_script("window.scrollTo(0, 400)")

        try:
            ele = driver.find_element(By.CSS_SELECTOR, '.e-1en57se').get_attribute('innerHTML')
            ele = ele.split(">")
            price = ele[1].split("<")[0]
        except:
            pass

        try:
            ele = driver.find_element(By.CSS_SELECTOR, '.e-1en57se').get_attribute('innerHTML')
            ele = ele.split(">")
            weight = ele[3].split("<")[0]
        except:
            pass
    except:
        weight = "None"
        price = "None"
    f = open("prices.txt","a")
    f.write(item+"\t"+weight+"\t"+price+"\n")
    f.close()
```
